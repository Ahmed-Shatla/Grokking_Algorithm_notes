```python
def QuickSort(arr):
  if len(arr)<2:
    return arr
  else:
    pivot = arr[0]
    left_sub_arr = [i for i in arr[1:] if i <= pivot]
    right_sub_arr = [i for i in arr[1:] if i > pivot]
    return QuickSort(left_sub_arr)+[pivot]+QuickSort(right_sub_arr)
```

```python
arr=[2,1,0,3,4]
print(QuickSort(arr))
```
