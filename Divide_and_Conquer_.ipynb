```python
def sum(arr):
  if arr == []:
    return 0
  return arr[0]+sum(arr[1:])
```

```python
arr=[1,2,3]
print(sum(arr))
```

```python
def count(arr):
  if arr == []:
    return 0
  return 1+count(arr[1:])
```

```python
arr=[1,2,133,44]
print(count(arr))
```

```python
def maxNum(arr):
  if len(arr)== 2:
    if arr[0]>arr[1]:
      return arr[0]
    else:
      return arr[1]
  else:
    return arr[0] if arr[0]>maxNum(arr[1:]) else maxNum(arr[1:])
```

```python
arr=[1,2,133,44]
print(maxNum(arr))
```
